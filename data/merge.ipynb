```python
import json

# 로드할 파일 리스트
file_names = ['classified_output_655_754.json', 'classified_output_755_854.json', 'classified_output_855_954.json', 'classified_output_955_1054.json']
combined_data = []

# 1. 4개의 파일 로드 및 합치기
for file_name in file_names:
    with open(file_name, 'r', encoding='utf-8') as f:
        data = json.load(f)
        combined_data.extend(data)  # 리스트 끝에 이어붙임

# 2. 결과 내보내기
with open('classified_influencers_result.json', 'w', encoding='utf-8') as f:
    json.dump(combined_data, f, ensure_ascii=False, indent=4)

print(f"총 {len(combined_data)}개의 아이템이 합쳐졌습니다!")
```

```python
json_file = 'classified_influencers_result.json'
jsonl_file = 'classification_debug_log_455_654.jsonl'
```

```python
import json
combined_data = []

# 1. JSON 파일 로드 (리스트 형태라고 가정)
with open(json_file, 'r', encoding='utf-8') as f:
    json_data = json.load(f)
    if isinstance(json_data, list):
        combined_data.extend(json_data)
    else:
        combined_data.append(json_data)

# 2. JSONL 파일 로드 (한 줄씩 읽어서 추가)
with open(jsonl_file, 'r', encoding='utf-8') as f:
    for line in f:
        if line.strip(): # 빈 줄 제외
            combined_data.append(json.loads(line))

# 3. 결과 내보내기 (JSON 형식으로 저장)
with open(json_file, 'w', encoding='utf-8') as f:
    json.dump(combined_data, f, ensure_ascii=False, indent=4)

print(f"✅ 합치기 완료! 총 {len(combined_data)}개의 데이터가 저장되었습니다.")
```
